```python
# CLass Coding Example

import requests #sends http request to github
import json #interprets instructions received back from github
from pprint import pprint

# get secret from Colab
from google.colab import userdata
token = userdata.get('ghtoken')

# initialize request parameters
# set url - api.github.com
url = 'https://api.github.com/user'
headers = {'Authorization': 'Bearer ' + token}

# send request to GitHub
r = requests.get(url, headers=headers)

# print initial responses
print(r.status_code)
print(r.text) #returns JSON file with user details requested as a string

#response 200 means that the request was successfully handled

r_json = json.loads(r.text)
pprint(r_json) #looks the same, but with a few format differences
```

```python
# Exercise: Explore the GitHub API

import requests #sends http request to github
import json #interprets instructions received back from github
from pprint import pprint

# get secret from Colab
from google.colab import userdata
token = userdata.get('ghtoken')

# set url - api.github.com
url = 'https://api.github.com/search/repositories?q=forks:>1&sort=forks&per_page=20'
headers = {'Authorization': 'Bearer ' + token}

# send request to GitHub
r = requests.get(url, headers=headers)

# print initial responses
print(r.status_code)
print(r.text) #returns JSON file with user details requested as a string

r_json = json.loads(r.text)
pprint(r_json) #looks the same, but with a few format differences
```

```python
items = r_json['items']
print(items[0])
```

```python
user_name = []
user_forks = []

for item in items:
  user_name.append(item['name'])
  user_forks.append(item['forks'])

print(user_name)
print(user_forks)
```

```python
import matplotlib.pyplot as plt

plt.barh(user_name, user_forks)
plt.title(label="Top 20 GitHub Users With The Most Forks")
plt.xlabel('Number of Forks')
plt.ylabel('User Names')
```
